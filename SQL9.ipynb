# CTE(Common table expression)

## Common Table Expressions (CTE) in SQL

1) What is CTE?

   * A Common Table Expression (CTE) is a temporary result set that you can reference within a SELECT, INSERT, UPDATE,          DELETE statement. It helps in improving the readability and maintainability of complex queries.

2) Why do we use CTE?
 
   *  We use CTEs to simplify complex queries, break them down into smaller, more manageable parts, and make them easier         to understand and maintain.

3) Three different use cases of CTE with code:
  
   * Recursive CTE to generate a sequence of numbers:
   
   ```sql
   WITH RECURSIVE numbers AS (
    SELECT 1 AS n
    UNION ALL
    SELECT n + 1 FROM numbers WHERE n < 10)
   SELECT * FROM numbers;
   ```
   
   * CTE for data transformation:
   
   ```sql
   WITH sales_total AS (
    SELECT 
        customer_id, 
        SUM(amount) AS total_amount
    FROM 
        sales
    GROUP BY 
        customer_id
   )
   SELECT * FROM sales_total;
   ```
   
   * CTE for hierarchical data:
   
   ```sql
   WITH recursive employee_hierarchy AS (
    SELECT 
        employee_id,
        manager_id,
        employee_name
    FROM 
        employees
    WHERE 
        manager_id IS NULL
    
    UNION ALL
    
    SELECT 
        e.employee_id,
        e.manager_id,
        e.employee_name
    FROM 
        employees e
    JOIN 
        employee_hierarchy eh ON e.manager_id = eh.employee_id
   )
   SELECT * FROM employee_hierarchy;
   ```
   
4) Where we can potentially use CTE:
   * Recursive queries to work with hierarchical data.
   * Data transformation and aggregation before joining with other tables.
   * Avoiding duplication of complex subqueries by defining them once in the CTE and then referencing them multiple times in the main query.
