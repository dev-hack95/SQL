```sql
DELIMITER $$
create function  mark_sales(sales decimal(20, 6))
returns varchar(30)
deterministic
begin
   declare flag_sales varchar(30);
   if sales <= 1000 then 
       set flag_sales = "super affordable";
   elseif sales > 1000 and sales <= 3000 then
       set flag_sales = "affordable" ;
   elseif sales > 3000 and sales <= 6000 then
       set flag_sales = "moderate";
   else
       set flag_sales = "expensive";
   end if;
   return flag_sales;
end $$
DELIMITER ;
```
